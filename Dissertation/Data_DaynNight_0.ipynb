```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("heonh0/daynight-cityview")

print("Path to dataset files:", path)
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

day_dir = os.path.join(path, "day")
night_dir = os.path.join(path, "night")

# List files inside day folder
day_files = os.listdir(day_dir)
night_files = os.listdir(night_dir)

day_img = Image.open(os.path.join(day_dir, day_files[1]))
night_img = Image.open(os.path.join(night_dir, night_files[1]))

# Plot them side by side
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].imshow(day_img)
axs[0].set_title("Day")
axs[0].axis("off")

axs[1].imshow(night_img)
axs[1].set_title("Night")
axs[1].axis("off")

plt.show()
```
