```python
import os
import shutil
import random

# Paths
data_dir = "/content/drive/MyDrive/MAI/Dissertation/CycleGAN_erik_linder/data/day2night"
output_dir = "/content/drive/MyDrive/MAI/Dissertation/CycleGAN_erik_linder/data/day2night"

day_dir = os.path.join(data_dir, "day")
night_dir = os.path.join(data_dir, "night")

# Train/test split ratio
train_ratio = 0.8

# Create output folders
for split in ["train", "test"]:
    for subfolder in ["A", "B"]:  # A = day, B = night
        os.makedirs(os.path.join(output_dir, split, subfolder), exist_ok=True)

def split_and_copy(src_dir, dest_dir_a, dest_dir_b, ratio=0.8):
    files = [f for f in os.listdir(src_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    random.shuffle(files)
    split_idx = int(len(files) * ratio)

    # Train files
    for f in files[:split_idx]:
        shutil.copy(os.path.join(src_dir, f), os.path.join(dest_dir_a if 'day' in src_dir else dest_dir_b, f))
    # Test files
    for f in files[split_idx:]:
        shutil.copy(os.path.join(src_dir, f), os.path.join(dest_dir_b if 'night' in src_dir else dest_dir_a, f))

# Copy day images to train/test A folders
day_train_dir = os.path.join(output_dir, "train", "A")
day_test_dir = os.path.join(output_dir, "test", "A")
split_and_copy(day_dir, day_train_dir, day_test_dir, ratio=train_ratio)

# Copy night images to train/test B folders
night_train_dir = os.path.join(output_dir, "train", "B")
night_test_dir = os.path.join(output_dir, "test", "B")
split_and_copy(night_dir, night_train_dir, night_test_dir, ratio=train_ratio)

print("Dataset split completed!")
```

```python
import os

def check_dataset_structure(base_dir):
    """
    Checks if train/test folders and A/B subfolders exist,
    and prints the number of images in each.
    """
    for split in ["train", "test"]:
        for subfolder in ["A", "B"]:
            folder_path = os.path.join(base_dir, split, subfolder)
            if os.path.exists(folder_path):
                num_files = len([f for f in os.listdir(folder_path)
                                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
                print(f"{folder_path}: exists, contains {num_files} image(s)")
            else:
                print(f"{folder_path}: MISSING!")

# Example usage:
output_dir = "/content/drive/MyDrive/MAI/Dissertation/CycleGAN_erik_linder/data/day2night"
check_dataset_structure(output_dir)
```
